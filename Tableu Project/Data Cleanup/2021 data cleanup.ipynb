```python
import pandas as pd
import numpy as np
```

```python
#import csv for 2020
csv = "JC-202101-citibike-tripdata.csv"
```

```python
#read 2020 csv
Bike_trip_jan21 = pd.read_csv(csv)
```

```python
#renaming dataframes
Bike_trip_jan21_df = Bike_trip_jan21
```

```python
#vieweing dataframe 
Bike_trip_jan21_df
```

```python
#checking type of data in dataframe
Bike_trip_jan21_df.dtypes
```

```python
Bike_2021_df=Bike_trip_jan21_df
```

```python
#removing columns
del Bike_2021_df['usertype']
```

```python
# Changing The Type of Values for the Gender Column to reassign genders
Bike_2021_df['gender'] = Bike_2021_df['gender'].astype(str)
```

```python
# Changing the Numeric Gender Values to Actual Gender Values For The 2020 df
Gender_check_2021 = Bike_2021_df['gender']
Changed_2021_male = Gender_check_2021.replace('1', 'Male')
Changed_2021_female = Changed_2021_male.replace('2', 'Female')
Changed_2021_unknown = Changed_2021_female.replace('0', 'Unknown')
Bike_2021_df['gender'] = Changed_2021_unknown
Bike_2021_df
```

```python
# Calculating Biker's Age's 
current_year = 2024

Birth_year_2021 = Bike_2021_df['birth year']
age = current_year - Birth_year_2021
Bike_2021_df['age'] = age
```

```python
# Rearranging the Columns
Bike_2021_df_organized = Bike_2021_df[["bikeid",
                                       "gender",
                                       "birth year",
                                       "age",
                                       "tripduration",
                                       "start station id",
                                       "start station name",
                                       "start station latitude",
                                       "start station longitude",
                                       "starttime",
                                       "end station id",
                                       "end station name", 
                                       "end station latitude",
                                       "end station longitude",
                                       "stoptime"]]
Bike_2021_df_organized
```

```python
# Renaming the Columns for consistency
Bike_2021_renamed_df = Bike_2021_df_organized.rename(columns={"bikeid":"Bike ID", 
                                                              "gender":"Gender",
                                                              "birth year":"Birth Year",
                                                              "age":"Age",
                                                              "tripduration":"Trip Duration",
                                                              "start station id":"Start Station ID",
                                                              "start station name":"Start Station Name",
                                                              "start station latitude":"Start Station Latitude",
                                                              "start station longitude":"Start Station Longitude",
                                                              "starttime":"Start Time",
                                                              "end station id":"End Station ID",
                                                              "end station name":"End Station Name",
                                                              "end station latitude":"End Station Latitude",
                                                              "end station longitude":"End Station Longitude",
                                                              "stoptime":"Stop Time"})
Bike_2021_renamed_df.dropna
```

```python
# Outputting file to csv 
Bike_2021_renamed_df.to_csv("2021_citibike_tripdata_cleaned.csv", index=False, header=True)
```
