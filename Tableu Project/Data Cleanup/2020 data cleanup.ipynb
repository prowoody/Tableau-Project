```python
import pandas as pd
import numpy as np
```

```python
#import csv for 2020
csv1 = "JC-202001-citibike-tripdata.csv"
csv2 = "JC-202006-citibike-tripdata.csv"
csv3 = "JC-202012-citibike-tripdata.csv"
```

```python
#read 2020 csv
Bike_trip_jan20 = pd.read_csv(csv1)
Bike_trip_jun20 = pd.read_csv(csv2)
Bike_trip_dec20 = pd.read_csv(csv3)
```

```python
#renaming dataframes
Bike_trip_jan20_df = Bike_trip_jan20
Bike_trip_jun20_df = Bike_trip_jun20
Bike_trip_dec20_df = Bike_trip_dec20
```

```python
#vieweing dataframe 
Bike_trip_jan20_df
```

```python
#vieweing dataframe 
Bike_trip_dec20_df
```

```python
#checking type of data in dataframe
Bike_trip_dec20_df.dtypes
```

```python
#Merging jan and june 2020 dataframes
First_Merged_df = pd.merge(Bike_trip_jan20_df,Bike_trip_jun20_df, how = 'outer')
```

```python
#Vieweing dataframe
First_Merged_df 
```

```python
#Merging the merged dataframe on dec 2020 df to create one df for the year
Second_Merged_df = pd.merge(First_Merged_df,Bike_trip_dec20_df, how = 'outer')
```

```python
#renaming df and viewing it
Bike_2020_df = Second_Merged_df 
Bike_2020_df
```

```python
#removing columns
del Bike_2020_df['usertype']
```

```python
# Changing The Type of Values for the Gender Column to reassign genders
Bike_2020_df['gender'] = Bike_2020_df['gender'].astype(str)
```

```python
# Changing the Numeric Gender Values to Actual Gender Values For The 2020 df
Gender_check_2020 = Bike_2020_df['gender']
Changed_2020_male = Gender_check_2020.replace('1', 'Male')
Changed_2020_female = Changed_2020_male.replace('2', 'Female')
Changed_2020_unknown = Changed_2020_female.replace('0', 'Unknown')
Bike_2020_df['gender'] = Changed_2020_unknown
Bike_2020_df
```

```python
# Calculating Biker's Age's 
current_year = 2024

Birth_year_2020 = Bike_2020_df['birth year']
age = current_year - Birth_year_2020
Bike_2020_df['age'] = age
```

```python
# Rearranging the Columns
Bike_2020_df_organized = Bike_2020_df[["bikeid",
                                       "gender",
                                       "birth year",
                                       "age",
                                       "tripduration",
                                       "start station id",
                                       "start station name",
                                       "start station latitude",
                                       "start station longitude",
                                       "starttime",
                                       "end station id",
                                       "end station name", 
                                       "end station latitude",
                                       "end station longitude",
                                       "stoptime"]]
Bike_2020_df_organized
```

```python
# Renaming the Columns for consistency
Bike_2020_renamed_df = Bike_2020_df_organized.rename(columns={"bikeid":"Bike ID", 
                                                              "gender":"Gender",
                                                              "birth year":"Birth Year",
                                                              "age":"Age",
                                                              "tripduration":"Trip Duration",
                                                              "start station id":"Start Station ID",
                                                              "start station name":"Start Station Name",
                                                              "start station latitude":"Start Station Latitude",
                                                              "start station longitude":"Start Station Longitude",
                                                              "starttime":"Start Time",
                                                              "end station id":"End Station ID",
                                                              "end station name":"End Station Name",
                                                              "end station latitude":"End Station Latitude",
                                                              "end station longitude":"End Station Longitude",
                                                              "stoptime":"Stop Time"})
Bike_2020_renamed_df.dropna
```

```python
# Outputting file to csv 
Bike_2020_renamed_df.to_csv("2020_citibike_tripdata_cleaned.csv", index=False, header=True)
```
